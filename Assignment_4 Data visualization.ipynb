### Exploratory Data Analysis on data set 1

### 1 Barplot.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='room_type', y='price', data=df)
plt.xlabel('room_type',fontsize = 15)
plt.ylabel('price', fontsize = 15)
plt.title('Room Type vs Price Comparison')
plt.show()
```

#### Summary = Acording to the above graph we can see that Entire home/apt price is very expensive than others

### 2 Boxplot.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="minimum_nights",y="price",data = df,hue="neighbourhood_group")
plt.xlim(0,9)
plt.ylim(0,550)
plt.xlabel('minimum_nights')
plt.ylabel('price')
plt.title('minimum_nights with prices charged in neighbourhood_group')
plt.show()
```

#### Summary = From the above graph we can see that how much price is charged for minimum_nights by different group.

### 3 Violinplot.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='room_type', y='availability_365', data=df)
plt.xlabel('room_type')
plt.ylabel('availability_365')
plt.title('Room availability according to room type')
plt.show()
```

#### Summary = With the help of above graph we can see that how many rooms are available according to the room type.

### 4 Countplot

```python
ax = sns.countplot(x = 'Sex', data = df1)
```

```python
ax = sns.countplot(x = 'Sex', data = df1)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df1.groupby(['Sex'], as_index= False)['Salary'].sum().sort_values(by = 'Salary', ascending = False)
```

```python
sal_gen = df1.groupby(['Sex'], as_index= False)['Salary'].sum().sort_values(by = 'Salary', ascending = False)
sns.barplot(x = 'Sex',y = 'Salary', data = sal_gen)
```

### Summary = from above graphs we can see that most of the employees are females and even the salary of females greater than male. 

### 5 Pie Chart

```python
plt.figure(figsize=(8, 8))
plt.pie(df1['Department'].value_counts(), labels=df1['Department'].value_counts().index, autopct='%1.1f%%')
plt.title('Designation chart')
plt.show()
```

### Summary = from above graphs we can see that most of the employees are from Production department. 

### 6 Histogram plot

```python
grouped_data = df1.groupby('Department')['SpecialProjectsCount'].sum()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped_data.index, grouped_data.values, color='skyblue')
plt.xlabel('Department')
plt.ylabel('Special Project Count')
plt.title('Special Project Count by Department')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Summary = From the above graph we can see that most of Special Projects have been done by IT/IS Department.

### 7 Stacked Bar Plot

```python
data = df1.groupby('RecruitmentSource')['PerformanceScore'].value_counts().unstack().fillna(0)
data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('RecruitmentSource')
plt.ylabel('PerformanceScore')
plt.title('Performance Score According Recruitment Source')
plt.show()
```

### Summary = From the above graph we can see that the best performance employee has come from google search recruitetment source.

### 8 Scatter Plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(df1['Department'], df1['Absences'], alpha=0.5)
plt.xlabel('Department')
plt.ylabel('Absence')
plt.title('Absence by Department')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Summary = From above graph we can see that most of Absences are from Production Department.

### 9 Area plot

```python
plt.figure(figsize=(10, 6))
for department, data in df1.groupby('Sex'):
    plt.fill_between(data.index, data['Department'], label=department)

plt.xlabel('Sex')
plt.ylabel('Department')
plt.title('Gender Distribution Across Departments')
plt.legend()
plt.show()
```

#### Summary = From the above graph we can see that there are more females employee are working in every department than males.

```python
df1.columns
```

### 10 Heatmap plot

```python
data = df1.pivot_table(index='Salary', columns='Department', aggfunc='size', fill_value=0)
plt.figure(figsize=(10, 6))
sns.heatmap(data, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Salary Distribution Across Departments')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()
```

#### Summary = From the above graph we come to know that the company is spending maximum salary on the production department.

#### Conclusion

#### Summary = From the above visualization we can see that the company is spending maximum salary on the production department.
